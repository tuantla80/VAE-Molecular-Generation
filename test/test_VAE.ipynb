#### Get train loss from logfile

```python
import matplotlib.pyplot as plt
train_loss_avg = []
file_path = r'C:\Python\VAE\test\test_train_VAE.txt'
with open(file_path) as f:
    for line in f.readlines():
        if 'Average train loss of this epoch =' in line:
            loss = float(line.split('=')[1])
            train_loss_avg.append(loss)
```

#### Plot Epoch vs. Training loss

```python
%matplotlib notebook

fig = plt.figure()
train_loss = train_loss_avg
epochs = range(0, len(train_loss_avg))
plt.plot(epochs, train_loss, 'ko', label='Training loss')
plt.axhline(y=0.0, color='b', linestyle='--')
plt.title('Epoch vs. Training loss')
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.legend()
fig.savefig('Epoch vs. Training loss.png', dpi = 150, bbox_inches = 'tight')
# plt.show()
```
